# Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.   
Created by John D Hunter.

```python
import matplotlib
```

```python
matplotlib.__version__
```

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.array([0,6])
y=np.array([0,50])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,50])
plt.plot(x,y,'ro')
plt.show()
```

```python
x=np.array([0,2,3,6])
y=np.array([0,10,20,50])
plt.plot(x,y,marker='o',ls='-.',color='r')
plt.show()
```

```python
x=np.array([0,2,3,6])
y=np.array([0,10,20,50])
plt.plot(x,y,'o-.r')
plt.show()
```

```python
x=np.array([0,2,3,5])
y=np.array([0,30,10,50])
plt.plot(x,y,marker='o',ms=20,mfc='r',mec='b')
plt.show()
```

```python
x=np.array([0,2,3,5])
y=np.array([0,30,10,50])
plt.plot(x,y,marker='o',mfc='b',ms=10,mec='w',ls='-.',lw=3,color='r')
plt.show()
```

```python
x1=np.array([0,2,3,5])
y1=np.array([0,30,10,50])
x2=np.array([0,1.5,3.5,5.5])
y2=np.array([0,20,5,40])
plt.plot(x1,y1,marker='o',mfc='b',ms=10,mec='w',ls='-.',lw=3,color='r')
plt.plot(x2,y2,marker='+',mfc='y',ms=10,mec='b',ls=':',lw=3,color='g')
plt.show()
```

```python
x=np.array([0,2,3,5])
y=np.array([0,30,10,50])
plt.plot(x,y,marker='o',mfc='b',ms=10,mec='w',ls='-.',lw=3,color='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph',loc='left')
plt.show()
```

```python
x=np.array([0,2,3,5])
y=np.array([0,30,10,50])
plt.plot(x,y,marker='o',mfc='b',ms=10,mec='w',ls='-.',lw=3,color='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
plt.grid(axis='x',ls='-.',lw=1,color='g')
plt.show()
```

```python
plt.subplot(1,2,1)
x1=np.array([0,2,3,5])
y1=np.array([0,30,10,50])
plt.plot(x1,y1,marker='o',mfc='b',ms=10,mec='w',ls='-.',lw=3,color='r')
plt.subplot(1,2,2)
x2=np.array([0,1.5,3.5,5.5])
y2=np.array([0,20,5,40])
plt.plot(x2,y2,marker='+',mfc='y',ms=10,mec='b',ls=':',lw=3,color='g')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y,color='r',s=100,alpha=0.5)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,color='r',s=100,alpha=0.5)


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,color='b',s=100,alpha=0.5)

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,color='r',width=0.3)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y,color='r',height=0.3)
plt.show()
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x,bins=10)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,startangle=180,autopct='%0.2f%%',explode=[0.2,0,0,0],shadow=True)
plt.show() 
```
